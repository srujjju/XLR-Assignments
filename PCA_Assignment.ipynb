```python
import pandas as pd
file = pd.read_csv('wine.csv')
file.head()
print(file.columns)
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
SS_X = SS.fit_transform(file)
SS_X = pd.DataFrame(SS_X,columns= file.columns)
SS_X.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pc = pca.fit_transform(SS_X)
imp_df = pd.DataFrame(pc )
imp_df.head()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt


pca = PCA()
pca.fit(SS_X)


explained_variance_ratio = pca.explained_variance_ratio_
print(explained_variance_ratio)

# Plot explained variance ratio
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o', linestyle='-')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance Ratio by Number of Components')
plt.grid(True)
plt.show()
```

```python
from sklearn.decomposition import PCA

desired_number_of_components = 3

pca = PCA(n_components=desired_number_of_components)
SS_X_reduced = pca.fit_transform(SS_X)
SS_X_PCA = pd.DataFrame(SS_X_reduced,columns=['PC1','PC2','PC3'])
SS_X_PCA.head()
```

```python

import matplotlib.pyplot as plt
# Hierarchical Clustering

from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')

labels_hc = hc.fit_predict(SS_X_PCA)

print('Hierarchical clustering labels:')
print(labels_hc)

# **K-Means Clustering**


from sklearn.cluster import KMeans

km = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)


labels_km = km.fit_predict(SS_X_PCA)

print('K-Means clustering labels:')
print(labels_km)

# **Elbow Curve**

from sklearn.metrics import silhouette_score

# Calculate silhouette scores for different numbers of clusters
silhouette_scores = []
for n_clusters in range(2, 15):
    clusterer = KMeans(n_clusters=n_clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)
    clusterer.fit(SS_X_PCA)
    silhouette_scores.append(silhouette_score(SS_X_PCA, clusterer.labels_))

# Plot the elbow curve
plt.figure(figsize=(10, 6))
plt.plot(range(2, 15), silhouette_scores, marker='o', linestyle='-')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Elbow Curve for K-Means Clustering')
plt.grid(True)
plt.show()


```
